```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo ponderado
G = nx.Graph()
G.add_weighted_edges_from([(1, 2, 2), (1, 3, 4), (2, 3, 1), (2, 4, 7), (3, 4, 3)])

# Calcular el camino más corto desde el nodo 1
shortest_paths = nx.single_source_dijkstra_path_length(G, 1)

# Visualizar el grafo
pos = nx.spring_layout(G)
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, edge_color='gray')

# Dibujar las distancias más cortas como etiquetas
for node, length in shortest_paths.items():
    plt.text(pos[node][0], pos[node][1] + 0.05, f'{length}', fontsize=12, ha='center')

plt.title("Dijkstra: Distancias más cortas desde el nodo 1")
plt.show()

```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo ponderado
G = nx.Graph()
G.add_weighted_edges_from([(1, 2, 1), (1, 3, 4), (2, 3, 2), (2, 4, 5), (3, 4, 1)])

# Encontrar el árbol de expansión mínima usando Kruskal
mst = nx.maximum_spanning_tree(G)

# Visualizar el grafo original y el MST
pos = nx.spring_layout(G)
plt.figure(figsize=(12, 8))

# Grafo original
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, edge_color='gray', alpha=0.5, style='dotted')

# Aristas del MST
nx.draw_networkx_edges(G, pos, edgelist=mst.edges(), edge_color='r', width=2)

plt.title("Árbol de Expansión Mínima (Kruskal)")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Crear un grafo ponderado
G = nx.Graph()
G.add_weighted_edges_from([(1, 2, 1), (1, 3, 4), (2, 3, 2), (2, 4, 7), (3, 4, 3)])

# Calcular todos los caminos más cortos
shortest_paths = dict(nx.all_pairs_dijkstra_path_length(G))

# Crear una matriz de distancias
nodes = list(G.nodes())
dist_matrix = np.zeros((len(nodes), len(nodes)))

for i, u in enumerate(nodes):
    for j, v in enumerate(nodes):
        dist_matrix[i, j] = shortest_paths[u].get(v, float('inf'))

# Graficar la matriz de distancias
plt.figure(figsize=(8, 6))
plt.imshow(dist_matrix, cmap='viridis', interpolation='none')
plt.colorbar(label='Distancia')
plt.xticks(ticks=np.arange(len(nodes)), labels=nodes)
plt.yticks(ticks=np.arange(len(nodes)), labels=nodes)
plt.title("Matriz de Distancias (Floyd-Warshall)")
plt.show()
```
